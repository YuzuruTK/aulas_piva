```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# Erro Amostral Toleravel, Normalmente não passa dos 5%
e = int(input("Diga o Erro Amostral Tolaravel em (sem o %): "))
# Tamanho da População
N = int(input("Diga o Tamanho da população: "))

# Execução do Código
tolerable_error = e/100
# Tamanho Minimo da Amostra
first_aproximation = (1/pow(tolerable_error,2)).__ceil__()
print(f"Tamanho da Minimo da amostra = {first_aproximation}")

amostral_size = float((N*first_aproximation)/(N+first_aproximation)).__ceil__()
print(f"Correção da amostra: {amostral_size}")
print(f"{amostral_size} é {(amostral_size/N)*100}% da população")
```

```python
def calculate_sample(tolerable_error: int, population_size: int):
    # Execução do Código
    tolerable_error = tolerable_error/100
    # Tamanho Minimo da Amostra
    first_aproximation = (1/pow(tolerable_error,2)).__ceil__()
    amostral_size = float((population_size*first_aproximation)/(population_size+first_aproximation)).__ceil__()
    return amostral_size
def graphic_generation():
    populacion = {"População" : [], "Amostra" : []}
    for i in range(500):
        x = calculate_sample(e, (i+1)*N)
        populacion["População"].append((i+1)*N)
        populacion["Amostra"].append(x)
    # print(populacion)
    dataframe = pd.DataFrame(data=populacion)
    # print(dataframe)
    sns.lineplot(data=dataframe,x="População", y= "Amostra")
    plt.show()
graphic_generation()
```
